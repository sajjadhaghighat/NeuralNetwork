```
import tensorflow as tf
from sklearn import preprocessing

data = tf.keras.datasets.boston_housing

(x_train, y_train),(x_test, y_test) = data.load_data()
```

```
scaler = preprocessing.StandardScaler( )
x_train = scaler.fit_transform( x_train )
x_test = scaler.fit_transform( x_test )
```

```
model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(52, activation='sigmoid',input_shape=(13,)),
  tf.keras.layers.Dense(39, activation='sigmoid'),
  tf.keras.layers.Dense(26, activation='sigmoid'),
  tf.keras.layers.Dense(13, activation='sigmoid'),
  tf.keras.layers.Dense(1)
])
```

```
model.compile(optimizer = tf.keras.optimizers.Adagrad(learning_rate=0.5),
              loss='mean_squared_error',
              metrics=['mse'])
```

```
model.fit(x_train, y_train, epochs=5000, shuffle=True)
```

```
model.evaluate(x_test,y_test)
```

```
Y = model.predict(x_test)
for i in range(len(Y)):
  print("Actual :", y_test[i],"\tPredict : ", Y[i][0])
  print('___________________________________________')
```
